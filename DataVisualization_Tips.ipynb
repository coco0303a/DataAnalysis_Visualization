# 4.Data Visualization

## Basic Plots for EDA

### Line Plot

```python
# Set the width and height of the figure
plt.figure(figsize=(10,4))

# Add title
plt.title("Total bill by day")

# Line plot showing average total_bill by day
sns.lineplot(data=tips, x='day', y='total_bill', ci=None)
```

```python
# Line Plot 에 hue를 통해 month 값에 따라 여러 색으로 나눠 그리기 

# 그래프 그리기
sns.lineplot(data=tips, x="day", y="total_bill", hue="sex", ci = None)   

# legend 그래프 밖에 표시
plt.legend(bbox_to_anchor=(1, 1))                                        
```

* relplot : kind="line" 으로 하면 line plot을 그려준다
  > relplot (kind: 'scatter'; default or 'line')
* seaborn > Tutorial > Overview of seaborn plotting functions 
  > https://seaborn.pydata.org/tutorial/function_overview.html
* seaborn > API > seaborn.relplot
  > https://seaborn.pydata.org/generated/seaborn.relplot.html

```python
# relplot로 line plot 형태로 2개의 sub plot 을 그릴 수 있다  (kind에 line을 입력)
sns.relplot(data=tips, x='day', y='total_bill', kind='line', ci=None, hue='sex', col = 'sex', col_wrap=2)
```

```python
sns.relplot(x='day', y='tip', kind='line', col='smoker', hue='sex', ci=None, data=tips)
```

```python
# Point Plot
sns.pointplot(data=tips, x="day", y="tip", ci= None)  
```

### Bar Plot

```python
# Bar chart showing average total bill by day
sns.barplot(data=tips, x="day", y="total_bill", errwidth=0)
```

```python
# Bar chart showing average tips by size
sns.barplot(data=tips, x="size", y="tip", errwidth=0)
```

```python
# Horizontal bar chart
sns.barplot(data=tips, x="tip", y="size", errwidth=0, orient='h')
```

```python
# Bar chart showing average total bill by day & smoking
sns.barplot(data=tips, x="day", y="total_bill", errwidth=0, hue='smoker')
plt.legend(bbox_to_anchor=(1, 1))    
```

```python
# Count Plot showing Count of observation by day 
sns.countplot(data=tips, x='day')
```

```python
# Count Plot showing Count of observation by day and time
sns.countplot(data=tips, x='day', hue='time')
```

### Box Plot

```python
# Box Plot displaying min, first quartile [Q1], median, third quartile [Q3], max and outliers of total bill
sns.boxplot(data=tips, x='total_bill')
```

```python
# Box Plot displaying min, Q1, median,Q3, max , outliers of total bill by day
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
# Box Plot displaying min, Q1, median,Q3, max , outliers  of tip by day and gender
sns.boxplot(x='day', y='tip', hue='sex', data=tips, linewidth=2.5)
```

```python
# Box Plot displaying min, Q1, median,Q3, max , outliers of total bill by day and gender
sns.boxplot(x='day', y='total_bill', hue='sex', data=tips, linewidth=2.5)
```

### Scatter Plot

```python
# Scatter Plot
# 우리는 해당 그래프로 레스토랑에서 전체 식사비용(total_bill)에 따른 팁(tip)을 얼마나 지불했는지를 파악할 수 있습니다
# hue 에 입력되는 값이 명목형이 아닌 실수형일 경우, sequencial palette로 표현해준다
# color palettes - https://seaborn.pydata.org/tutorial/color_palettes.html

sns.scatterplot(data=tips, x="total_bill", y="tip" , hue='size', size='size', sizes=(15, 200), alpha=.9)
```

```python
# scatter plot  - time을 기준으로
# hue를 사용하여 day로 세분화
# 데이터를 하나 더 추가하여 더욱 세분화 해서 표현하고 싶으시다면 style 파라미터를 이용한다

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", style="time")  
```

```python
# scatter plot
#  style argument  - marker style 을 의미

sns.scatterplot(x="total_bill", y="tip", hue="smoker", style="smoker", data=tips)
```

```python
# Scatter Plot - day를 기준으로
# sunday의 성별에 따른 tip과 total_bill의 관계  -> femail의 경우 total_bill 대비 tip의 비율이 높다!

plt.figure(figsize=(10, 10))
tips_sun = tips.query('day == "Sun"')
sns.scatterplot(data=tips_sun, x="total_bill", y="tip", hue="sex"  ) 
```

## More Plots

### Hist Plot

```python
# His Plot
# 변수에 대한 히스토그램을 표시한다
# 하나 혹은 두 개의 변수 분포를 나타내는 전형적인 시각화 도구로 범위에 포함화는 관측수를 세어 표시한다.

sns.histplot(data=tips, x='total_bill' )
```

```python
tips_female = tips.query('sex == "Female"')
sns.histplot(data=tips_female, x='total_bill')
```

```python
tips_Male = tips.query('sex == "Male"')
sns.histplot(data=tips_Male, x='total_bill')
```

* displot : kind="hist" 하거나 default로 hist plot을 그려준다
> histplot (kind: 'hist'; default or 'kde')

* seaborn > Tutorial > Overview of seaborn plotting functions
> https://seaborn.pydata.org/tutorial/function_overview.html

* seaborn > API > seaborn.displot
> https://seaborn.pydata.org/generated/seaborn.displot.html



```python
sns.displot(data=tips, x='total_bill' , col= 'sex')
```

### Kde Plot

```python
# KDE Plot
# 하나 혹은 두 개의 변수에 대한 분표를 그린다.
# histplot은 절대량이라면 kdeplot은 밀도 추정치를 시각화한다
# 그래서 결과물로는 연속된 곡선의 그래프를 얻을 수 있다

sns.kdeplot(data=tips, x=tips['total_bill'])
```

```python
# distplot = kde plot + hist plot

sns.distplot(tips["total_bill"])
```

```python
# kde plot

sns.kdeplot(data=tips, x= 'size', hue='sex')
```

### Joint Plot

```python
# Joint Plot
# 두 변수에 대한 displot의 조합이다.
# 두 변수에 분포에 대한 분석을 할 수 있다
# 두 displot 사이에 scatter plot이 추가되어 분포를 추가로 확인할 수 있다.

sns.jointplot(data=tips, x='total_bill', y='tip', kind = 'scatter')
```

### Pair Plot

```python
# Pair Plot
# 데이터셋을 통째로 넣으면 숫자형 특성에 대하여 각각에 대한 히스토그램과 두 변수 사이의 scatter plot을 그린다.
sns.pairplot(tips)
```

```python
sns.pairplot(data=tips, hue="sex")
```

### Reg Plot

```python
# Reg Plot
# Regression 결과를 그래프로 보여준다.
# egression plot은 지금까지 봐온 plot과는 개념이 조금 다른데, regression 머신러닝의 결과를 그래프에 표기해준다.
# 즉 데이터를 있는 그대로 보여주기보다는, 머신러닝 결과와 함께 보여주는 것이다.

sns.regplot(x = 'tip', y = 'total_bill', data = tips)
```
